Importing Necessary Libraries

```python
import pandas as pd
import json
```

Loading the file paths

```python
csv_file_path = 'sales_csv.csv'
json_file_path = 'sales_json.json'
excel_file_path = 'sales_excel.xlsx'
```

Loading the data

csv

```python
csv_data = pd.read_csv(csv_file_path)
```

```python
csv_data
```

json

```python
with open(json_file_path, 'r') as file:
    json_data = json.load(file)
    
json_data = pd.json_normalize(json_data)
```

```python
json_data
```

excel

```python
excel_data = pd.read_excel(excel_file_path)
```

```python
excel_data
```

Preprocessing

```python
# normalizing
def normalize_columns(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    return df

csv_data = normalize_columns(csv_data)
json_data = normalize_columns(json_data)
excel_data = normalize_columns(excel_data)
```

```python
def convert_dates(df):
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    return df

csv_data = convert_dates(csv_data)
json_data = convert_dates(json_data)
excel_data = convert_dates(excel_data)

def convert_sales(df):
    df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
    return df

csv_data = convert_sales(csv_data)
json_data = convert_sales(json_data)
excel_data = convert_sales(excel_data)
```

```python
# Combining the all the data
all_data = pd.concat([csv_data, json_data, excel_data], ignore_index=True)
```

```python
# Save the combined data into a file
all_data.to_csv('unified_sales_data.csv', index=False)
```
